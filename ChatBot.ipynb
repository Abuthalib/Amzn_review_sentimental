```python
import speech_recognition as sr

class ChatBot:
    def __init__(self, name):
        self.name = name
        self.text = ""
    
    def speech_to_text(self):
        recognizer = sr.Recognizer()
        with sr.Microphone() as mic:
            print("listening...")
            audio = recognizer.listen(mic)
        try:
            self.text = recognizer.recognize_google_cloud(audio)
            print("me --> ", self.text)
        except Exception as e:
            print("me --> ERROR")
            print(str(e))

if __name__ == "__main__":
    ai = ChatBot(name="Dev")
    while True:
        ai.speech_to_text()
```
